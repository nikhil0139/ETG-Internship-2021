### 1. In how many cities Summer Olympics is held so far?

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
import pandas as pd
ans2 = pd.read_csv('summer.csv')

# grouping sports according to most number of gold medals and sort in descending order
ans2 = ans2[ans2['Medal'] == 'Gold'].groupby('Sport').count()['Medal'].sort_values(ascending = False).head()

# plot the graph of the values
ans2.plot.bar()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
import pandas as pd
ans3 = pd.read_csv('summer.csv')

# grouping sports according to most number of medals and sort in descending order
ans3 = ans3.groupby('Sport').count()['Medal'].sort_values(ascending = False).head()

# plot the graph of the values
ans3.plot.bar()
```

### 4. Which player has won most number of medals? (Top 5)

```python
import pandas as pd
ans4 = pd.read_csv('summer.csv')

# grouping players according to most number of medals and sort in descending order
ans4 = ans4.groupby('Athlete').count()['Medal'].sort_values(ascending = False).head()

# plot the graph of the values
ans4.plot.bar()
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
import pandas as pd
ans7 = pd.read_csv('summer.csv')

# grouping events according to most number of players and sort in descending order
ans7 = ans7.groupby('Event').count()['Athlete'].sort_values(ascending = False).head()

# plot the graph of the values
ans7.plot.bar()
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
import pandas as pd
ans8 = pd.read_csv('summer.csv')

# filtering values in which Gender is Women
ans8 = ans8[ans8['Gender'] == 'Women']

# grouping sports according to most number of gold medals in the data and sort in descending order
ans8 = ans8[ans8['Medal'] == 'Gold'].groupby('Sport').count()['Athlete'].sort_values(ascending = False).head()

# plot the graph of the values
ans8.plot.bar()
```
